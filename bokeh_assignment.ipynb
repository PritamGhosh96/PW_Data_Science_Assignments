**Q1. How can you create a Bokeh plot using Python code?**

```python
import bokeh.io
import bokeh.plotting
bokeh.io.output_notebook()
```

```python
from bokeh.plotting import figure, output_file, show
```

```python
x = [3,4,5,6,2,8]
y = [4,3,5,1,8,9]

output_file("line.html")
p = figure(title = 'line graph')

p.scatter(x, y, size = 20, fill_color = 'red', legend_label = "this is my data point")
show(p)
```

**Q2. What are glyphs in Bokeh, and how can you add them to a Bokeh plot? Explain with an example.**

Glyphs are the basic visual building blocks of Bokeh plots. They are vectorized graphical shapes or markers that are used to represent data.
 For example, when you want to create a scatter plot, you may use a circle as a marker to represent information. A line will represent information on a line plot.

```python
from bokeh.plotting import figure, show
from bokeh.io import output_file

# prepare some data
x = [1, 2, 4, 6, 7]
y = [7, 6, 3, 9, 10]

# create a new plot with figure function
p = figure(title="Circle Glyph")

# create circle glyph
p.circle(x=x, y=y, size=25, fill_color="red")
p.circle_cross(x=[2, 4, 6, 8], y=[5, 8, 9, 11], size=25,
               fill_color="blue", fill_alpha=0.3, line_width=2)
p.circle_x(x=[4, 7, 2, 6], y=[7, 2, 4, 9], size=25,
           fill_color="green", fill_alpha=0.6, line_width=2)

# show the results
output_file('circle.html')
show(p)
```

**Q3. How can you customize the appearance of a Bokeh plot, including the axes, title, and legend?**

```python
x = [3,4,5,6,2,8]
y = [4,3,5,1,8,9]

output_file("line.html")
p = figure(title = 'line graph') # adding title
p.xaxis.axis_label = "X Axis"
p.yaxis.axis_label = "Y Axis" # axes customizing

p.scatter(x, y, fill_color = 'red', legend_label = "this is my data point") # adding lebel
show(p)
```

**Q4. What is a Bokeh server, and how can you use it to create interactive plots that can be updated in real time?**

A Bokeh server is a Python library that creates interactive web applications that connect front-end UI events to running Python code. Bokeh creates high-level Python models, such as plots, ranges, axes, and glyphs, and then converts these objects to JSON to pass them to its client library, BokehJS.
Bokeh can be used to create interactive visualizations for modern web browsers, including Jupyter Notebook and Refinitiv CodeBook. It can handle custom or specialized use cases very simply and provides high-performance interactive charts and plots.

To create interactive plots that can be updated in real time, you can use the Bokeh server to:

  1.Convert the Python code you write to a JSON document
  
  2.Render the JSON document in JavaScript using Bokeh's client library (BokehJS)

Bokeh output can be obtained in various mediums like a notebook, HTML, and server.

**Q5. How can you embed a Bokeh plot into a web page or dashboard using Flask or Django?**

1. Using the bokeh.embed.components() function:

This function returns the script and an HTML div tag assuming that the required JS files from Bokeh are already loaded in your template through the local URL or Bokeh's CDN. The charts will be responsive due to the attribute 'sizing_mode=stretch_width' used in each of the figures.

2. Using the bokeh.plotting.file_html() function:

This function returns a raw HTML code representing a Bokeh plot. You can then save this HTML file or use it to render through URL routes in Flask app.
