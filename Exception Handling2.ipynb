Q3. What errors are defined in the ArithmeticError class? Explain any two with an example.

------ ArithmeticError

--------- FloatingPointError

--------- OverflowError

--------- ZeroDivisionError


```python
try:
    10/0
except ZeroDivisionError as e:
    print(e)
```

```python
j = 5.0

try:
    for i in range(1, 1000):
        j = j**i
except ArithmeticError as e:
    print(f"{e}, {e.__class__}")
```

Q4. Why LookupError class is used? Explain with an example KeyError and IndexError.

LookupError(StandardError) is used as a base class for exceptions raised when a sequence or dictionary type doesn't contain a given index or key

```python
try:
    l = [1,2,3,4]
    l[10]
except IndexError as e:
    print(e)
```

```python
try:
    d = {1 : [2,3,4,5], "prit": "ghosh"}
    print(d["key10"])
except KeyError as e:
    print(e)
```

Q5. Explain ImportError. What is ModuleNotFoundError?

ImportError typically occurs when you try to import a library or module that Python cannot find or access

ModuleNotFoundError is a Python error that occurs when you try to import a module that is not installed or not found in the current Python path
