###### Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

Any five plots that we can plot using the Seaborn library : 
    
    1. scatterplot - to display the relationship between two variables and observe the nature of the relationship
    2. barplot - to compare the different sets of data among different groups easily
    3. violinplot - It shows peaks in the data. It is used to visualize the distribution of numerical data
    4. kdeplot - a method for visualizing the distribution of observations in a dataset, analogous to a histogram.
    5. displot - A histogram, shows the frequency of a given variable in a barplot.

###### Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue parameter for the 'cut' column of the diamonds dataset.
