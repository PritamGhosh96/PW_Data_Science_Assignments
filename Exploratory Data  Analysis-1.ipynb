###### Q3. What are the key factors that affect students' performance in exams? How would you go about analyzing these factors using statistical techniques?

```python
stud = pd.read_csv('stud.csv')
```

```python
stud.head()
```

```python
stud.isnull().sum()
```

```python
stud.duplicated().sum()
```

```python
stud['Avg'] = (stud['math_score']+stud['reading_score']+stud['writing_score'])/3
```

We created average score to get score for overall scores.

```python
stud.head()
```

Gender vs Scores

```python
plt.figure(figsize=(20,10))
import seaborn as sns
sns.kdeplot(data = stud, x='Avg',hue = 'gender')
```

Observation :-

- Female scores are better on average than male scores

Race Ethnicity vs Scores

```python
sns.kdeplot(data = stud, x='Avg',hue = 'race_ethnicity')
```

Observation :-

- On Average Group E people are doing better folloewed by group D and Group C.
- Group A and C people are lagging on the tests.

Parent's Education vs Scores

```python
sns.kdeplot(data = stud, x='Avg',hue = 'parental_level_of_education')
```

Observation :-

- Students having Parent's with master's degree are doing best while Students having Parent's with High school degree are doing worst.
- Parents Education have positive effect on their child.

Lunch vs Scores

```python
sns.kdeplot(data = stud, x='Avg',hue = 'lunch')
```

Observation :-

- Students having standard lunch having a higher score than student having free/reduced lunch.
- Student lunch have a positive effect on student's score.

Preparation vs Scores

```python
sns.kdeplot(data = stud, x='Avg',hue = 'test_preparation_course')
```

```python
stud['test_preparation_course'].value_counts()
```

Observation :-

- Most Students didn't completed test preparation course
- Students who have completed test preparation course tends to do better.
